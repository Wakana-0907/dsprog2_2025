```python
x = 1
y = 0
```

```python
# 0で割っているのでエラーが起きる
x / y
```

```python
try:
     result = x / y
except ZeroDivisionError as e:
     result = None
     print("エラーが発生しました:", e)   
     print("0で割ることはできません。結果をNoneに設定します。。")

print("結果:", result)
```

## 複数のエラーが発生する可能性がある場合

```python
try:
    result = x / z
except ZeroDivisionError as e:
    result = None
    print("エラーが発生しました:", e)   
    print("0で割ることはできません。結果をNoneに設定します。。")
except NameError as e:
    result = None
    print("エラーが発生しました:", e)   
    print("変数が定義されていません。結果をNoneに設定します。。")

print("結果:", result)
```

## エラーが発生しなかった場合にのみ実行したい処理がある場合

```python
z = 3

try:
    result = x / z
except ZeroDivisionError as e:
    result = None
    print("エラーが発生しました:", e)   
    print("0で割ることはできません。結果をNoneに設定します。。")
except NameError as e:
    result = None
    print("エラーが発生しました:", e)   
    print("変数が定義されていません。結果をNoneに設定します。。")
else:
    print("エラーは発生しませんでした。結果を表示します。")

print("結果:", result)
```

## エラー発生の有無に関わらず実行したい処理がある場合

```python
z = 0
try:
    result = x / z
except ZeroDivisionError as e:
    result = None
    print("エラーが発生しました:", e)   
    print("0で割ることはできません。結果をNoneに設定します。。")
except NameError as e:
    result = None
    print("エラーが発生しました:", e)   
    print("変数が定義されていません。結果をNoneに設定します。。")
else:
    print("エラーは発生しませんでした。結果を表示します。")
finally:
    print("計算処理が終了しました。")

print("結果:", result)
```

## システム終了以外の全てのエラーをハンドリングしたい場合

```python
try:
    result = x / y
except Exception as e:
    result = None
    print("予期せぬエラーが発生しました:", e)   
    print("計算できません。結果をNoneに設定します。。")
else:
    print("エラーは発生しませんでした。結果を表示します。")
finally:
    print("計算処理が終了しました。")

print("結果:", result)
```

## webスクレイピングでのエラー処理

```python
import requests
```

```python
url = 'https:www.musashino-u.ac.jp/invalid-page' # 存在しないURL
response = requests.get(url)
print("ステータスコード:", response.status_code)

# ステータスコードが200以外の例を発生させる
response.raise_for_status()  
```

```python
url = 'https:www.musashino-u.ac.jp/invalid-page' # 存在しないURL
try:
    response = requests.get(url)
    print("ステータスコード:", response.status_code)
    response.raise_for_status()  
except requests.exceptions.HTTPError as e:
    print("HTTPエラーが発生しました:", e)
except requests.exceptions.RequestException as e:
    print("リクエスト中にエラーが発生しました:", e)
else:
    print("webページに正常にアクセスできました。")
finally:
    print("webスクレイピング処理が終了しました。")
```
